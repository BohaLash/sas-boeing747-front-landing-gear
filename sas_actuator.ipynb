```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
K = 2e9
m = 30
A1 = 0.012272
A2 = 0.00591
V1 = 0.0000775
V2 = 0.0014
```

```python
Q = 5e-4
F = 1e5

t0 = 0
tf = 4
```

```python
h = 1e-6
```

```python
xpts = []
ypts = []

t = t0

x = 0
xd = 0
p1 = 1e5
p2 = 1e5
while t <= tf:
    
    xdd = (p1*A1 - p2*A2 + F)/m
    xd += xdd*h
    x += xd*h
    p1d = (Q - A1*xd)*K/(V1 + A1*x)
    p2d = (-Q + A2*xd)*K/(V2 - A2*x)
    p1 += p1d*h
    p2 += p2d*h

    t += h

    xpts.append(t)
    ypts.append(x)

print(t)
print(x)
```

```python
xpoints = np.array(xpts)
ypoints = np.array(ypts)

print(xpoints)

plt.plot(xpoints, ypoints)
plt.xlabel('time [s]')
plt.ylabel('actuator displacement [m]')
plt.show()
```
